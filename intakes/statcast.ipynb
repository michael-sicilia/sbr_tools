```python
import pybaseball as pbase
pybaseball.cache.enable()
COLUMNS = 'columns'
DATA_TYPE = 'data_type'
DATA = 'data'
METRICS = 'metrics'
OG_FILE = 'original_file_name'
PLAYER_TYPE = 'player_type'
```

```python
yrs = range(2015, 2026)


for yr in yrs:
    _json_list = []
    df_list = pbase.statcast(start_dt=f'{yr}-01-01', end_dt=f'{yr}-12-31')
    for i in range(0, len(df_list)): 
        data_json = {}
        data_json[DATA_TYPE] = f'statcast_{yr}_{i}'
        data_json[DATA] = df_list[i]
        data_json[COLUMNS] = list(df_list[i].columns)
        
```
