```python
import os 
import pandas as pd 

def loadCSVs(path):
    _dfs = []
    for file in os.listdir(path):
        _dict = {}
        if file.endswith('.csv'):
            full_path = os.path.join(path, file)
            df = pd.read_csv(full_path)
            _dict[file] = df
            _dfs.append(_dict)

    return _dfs
```

```python
COLUMNS = 'columns'
DATA_TYPE = 'data_type'
DATA = 'data'
METRICS = 'metrics'
OG_FILE = 'original_file_name'
PLAYER_TYPE = 'player_type'
```

```python
root = "c:\\"
path = root + 'retrosheet'
dfs = loadCSVs(path)
import intake

_json_list = []
for _dict in dfs: 
    for file_name, df in _dict.items():
        data_json = {}
        _file = file_name.split('_')[0]
        data_json[DATA_TYPE] = 'sabermetric'
        data_json[METRICS] = _file
        data_json[DATA] = df 
        data_json[COLUMNS] = list(df.columns)
        data_json[OG_FILE] = file_name

    _json_list.append(data_json)
```

```python
metric_scrubber = intake.MetricScrub(_json_list)
```

```python
map = metric_scrubber.standardizeColumnNames()
for _dict in _json_list:
    _dict[DATA] = df.rename(columns=map).to_json(orient='records')
```

```python
import json
with open('retrosheet.json', 'w') as file:
    json.dump(_dict, file, indent=4)
```
